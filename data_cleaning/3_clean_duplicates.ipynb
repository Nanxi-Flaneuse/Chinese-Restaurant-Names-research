## merge parallel data with annotated data points

```python
# sanity check if the names match
selectedNUM = 3399

df_ = pd.read_csv("utilities/labeled_datasets/en_golden_label.csv")
print(df_.national_id.unique())
print(df_[df_.national_id == selectedNUM][['English_Name', 'Chinese_Name']])
print(df_sample[df_sample.national_id == selectedNUM][['English_Name', 'Chinese_Name']])
```

```python
def merge(host_file, guest_file, method, column_name, save_file):
    host = pd.read_csv(host_file)[['national_id']]
    guest = pd.read_csv(guest_file)
    host = host.merge(guest, how=method, on=column_name)
    host = host.dropna(subset=['google_id'])
    host.to_csv(save_file, index=False, encoding='utf-8')
    return host

df_parallel_en = merge('utilities/labeled_datasets/cn_golden_label.csv',
      'output/sample_final_parallel.csv','left','national_id','output/cn_golden_label.csv')
print("English parallel", df_parallel_en.shape)

df_parallel_cn = merge('utilities/labeled_datasets/en_golden_label.csv',
      'output/sample_final_parallel.csv','left','national_id','output/en_golden_label.csv')
print("Chinese parallel", df_parallel_cn.shape)
```

```python
def split(df_, validation_filename, test_filename):
    df_ = df_.sample(frac=1.0)
    half_size = int(df_.shape[0] / 2)
    validation, test = df_[:half_size], df_[half_size:]
    validation.to_csv(validation_filename, index=False, encoding='utf-8')
    test.to_csv(test_filename, index=False, encoding='utf-8')
    print(validation.shape, test.shape)
    return validation, test

print("English parallel", df_parallel_en.shape)
df_parallel_en = df_parallel_en.drop_duplicates(subset=["English_Name"])
df_parallel_en_val, df_parallel_en_test = split(df_parallel_en, "./output/validation_en.csv", "./output/test_en.csv")
print("Post splitting -- English parallel", df_parallel_en.shape)

print("Chinese parallel", df_parallel_cn.shape)
df_parallel_cn = df_parallel_cn.drop_duplicates(subset=["Chinese_Name"])
df_parallel_cn_val, df_parallel_cn_test = split(df_parallel_cn, "./output/validation_cn.csv", "./output/test_cn.csv")
print("Post splitting -- Chinese parallel", df_parallel_cn.shape)
```

```python
## sanity check
def check_intersection(list1, list2):
    return len(set(list1).intersection(list2)), set(list1).intersection(list2)

# no English name intersection for the English split
print(check_intersection(df_parallel_en_val.English_Name.unique(), df_parallel_en_test.English_Name.unique()))

# no Chinese name intersection for the Chinese split
print(check_intersection(df_parallel_cn_val.Chinese_Name.unique(), df_parallel_cn_test.Chinese_Name.unique()))
```
